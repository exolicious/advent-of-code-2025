```python
parsed_input = INPUT.split(",")
sum = 0
def is_invalid_id(id):
    first_sequence = set()
    id_str = str(id)
    for i in range(0, len(id_str)):
        if id_str[i] in first_sequence:
            first_sequence_string = id_str[0:i]
            first_sequence_string_length = len(first_sequence_string)
            chunks = [id_str[j:j+first_sequence_string_length] for j in range(0, len(id_str), first_sequence_string_length)]
            if all(chunk == chunks[0] for chunk in chunks):
                return True
        else:
            first_sequence.add(id_str[i])
    return False
for range_string in parsed_input:
    bounds = range_string.split("-")
    first_id = int(bounds[0])
    last_id = int(bounds[1])
    for id in range(first_id, last_id + 1):
        if is_invalid_id(id):
            sum += id
print(sum)

```
