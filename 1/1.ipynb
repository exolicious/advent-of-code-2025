```python
DIAL_START = 50
DIAL_AMOUNT = 99
```

```python
INPUT = """L68
L30
R48
L5
R60
L55
L1
L99
R14
L82"""
```

```python
parsed_input = INPUT.splitlines()
print(parsed_input)
current_dial_position = DIAL_START
password = 0
for rotation in parsed_input:
    rotation_amount = int(rotation[1:])
    sign = 1
    if rotation[0] == "L":
        sign *= -1
    current_dial_position = (current_dial_position + rotation_amount * sign) % (DIAL_AMOUNT+1)
    if current_dial_position == 0:
        password += 1
print(password)
```
