```python
parsed_input = INPUT.splitlines()
current_dial_position = DIAL_START
password = 0
print(f"DIAL START {current_dial_position}")

for rotation in parsed_input:
    rotation_amount = int(rotation[1:])
    sign = 1
    
    if rotation[0] == "L":
        print(f"left rotation by {rotation_amount}")
        sign = -1
        if current_dial_position == 0:
            first_zero = DIAL_AMOUNT 
        else:
            first_zero = current_dial_position 
        
        if first_zero <= rotation_amount:
            crossings = 1 + (rotation_amount - first_zero) // DIAL_AMOUNT
            password += crossings
            print(f"crossed 0 {crossings} times, password is now {password}")
            
    elif rotation[0] == "R":
        print(f"right rotation by {rotation_amount}")
        if current_dial_position == 0:
            first_zero = DIAL_AMOUNT
        else:
            first_zero = DIAL_AMOUNT - current_dial_position
        
        if first_zero <= rotation_amount:
            crossings = 1 + (rotation_amount - first_zero) // DIAL_AMOUNT
            password += crossings
            print(f"crossed 0 {crossings} time, password is now {password}")
    
    current_dial_position = (current_dial_position + rotation_amount * sign) % DIAL_AMOUNT
    print(f"Neew dial position {current_dial_position}")

print(password)
```
