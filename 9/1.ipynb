```python

class QuadrantNode:
    def __init__(self, points):
        self.quadrants = [None, None, None, None]
        self.points = points
        if len(points) < 10:
            self.points = points
            return
        self.centroid = QuadrantNode.calculate_centroid(points)
        QuadrantNode.create_and_fill_quadrants(self)

    @staticmethod
    def calculate_centroid(points):
        centroid_x_sum = 0
        centroid_y_sum = 0
        for point in points:
            centroid_x_sum += point[0]
            centroid_y_sum += point[1]
        return (centroid_x_sum / len(points), centroid_y_sum / len(points))

    @staticmethod
    def create_and_fill_quadrants(root: "QuadrantNode"):
        q1_node_points = []
        q2_node_points = []
        q3_node_points = []
        q4_node_points = []
        for point in root.points:
            if point[0] >= root.centroid[0] and point[1] >= root.centroid[1]:
                q1_node_points.append(point)
            elif point[0] < root.centroid[0] and point[1] >= root.centroid[1]:
                q2_node_points.append(point)
            elif point[0] < root.centroid[0] and point[1] < root.centroid[1]:
                q3_node_points.append(point)
            else:
                q4_node_points.append(point)
        if len(q1_node_points) > 0:
            root.quadrants[0] = QuadrantNode(q1_node_points)
        if len(q2_node_points) > 0:
            root.quadrants[1] = QuadrantNode(q2_node_points)
        if len(q3_node_points) > 0:
            root.quadrants[2] = QuadrantNode(q3_node_points)
        if len(q4_node_points) > 0:
            root.quadrants[3] = QuadrantNode(q4_node_points)

    @staticmethod
    def find_candidate_points_in_opposing_quadrants(root, first_quadrant_index, second_quadrant_index):
        temp = []
        best_points_in_first_quadrant = QuadrantNode.find_best_candidate_points_of_quadrant(root, first_quadrant_index)
        best_points_in_second_quadrant = QuadrantNode.find_best_candidate_points_of_quadrant(root, second_quadrant_index)
        
        temp.extend(best_points_in_first_quadrant)
        temp.extend(best_points_in_second_quadrant)
        return temp
    
    @staticmethod
    def find_best_candidate_points_of_quadrant(root: "QuadrantNode", preferred_quadrant_index):
        if len(root.points) < 5:
            return root.points
        if root.quadrants[preferred_quadrant_index]: # best candidates, early return
            return QuadrantNode.find_best_candidate_points_of_quadrant(root.quadrants[preferred_quadrant_index], preferred_quadrant_index)
        
        #other quadrants
        temp = []
        for quadrant_index in range(0, 4):
            if quadrant_index == preferred_quadrant_index:
                continue
            if root.quadrants[quadrant_index]:
                temp.extend(QuadrantNode.find_best_candidate_points_of_quadrant(root.quadrants[quadrant_index], preferred_quadrant_index))
        return temp
```

```python

def calculate_area(x1, x2, y1, y2):
    return abs(x1 - x2 + 1) * abs(y1 - y2 + 1)
```
