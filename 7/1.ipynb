```python
import numpy as np
import sys
np.set_printoptions(threshold=sys.maxsize)
```

```python
lines = INPUT.splitlines()
matrix = []

for line in lines:
    row = []
    symbols = list(line)
    for symbol in symbols:
        row.append(symbol)
    matrix.append(row)
print(matrix)

start_index_x = lines[0].index("S")
already_hit_splitters = set()

stack = []
stack.append((1, start_index_x))

while len(stack) > 0:
    origin = stack.pop()
    origin_y = origin[0]
    origin_x = origin[1]
    for i in range(origin_y, len(matrix)):
        if matrix[i][origin_x] == "^":
            if (i, origin_x) not in already_hit_splitters:
                already_hit_splitters.add((i, origin_x))
                stack.append((i, origin_x - 1))
                stack.append((i, origin_x + 1))
            break
        else:
            matrix[i][origin_x] = "|"
print(len(already_hit_splitters))
print(np.matrix(matrix))

print((len(already_hit_splitters)-1) * 2)
```
