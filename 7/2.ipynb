```python
def count_paths(y, x, cache, matrix):
    if (y, x) in cache:
        return cache[(y, x)]
    
    for i in range(y, len(matrix)):
        if i >= len(matrix):
            cache[(y, x)] = 1
            return 1
        if matrix[i][x] == "^":
            result = count_paths(i, x - 1, cache, matrix) + count_paths(i, x + 1, cache, matrix)
            cache[(y, x)] = result
            return result
    
    cache[(y, x)] = 1
    return 1
```
