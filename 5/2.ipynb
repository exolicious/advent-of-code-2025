```python
class BoundaryNode:
    def __init__(self, value, left: "BoundaryNode", right: "BoundaryNode"):
        self.value = value
        self.left = left
        self.right = right

    @staticmethod
    def insert(root: "BoundaryNode", left_bound: int, right_bound: int):
        if left_bound > root.right.value + 1:
            if root.left is None:
                new_right = BoundaryNode(right_bound, None, None)
                new_left = BoundaryNode(left_bound, None, new_right)
                root.left = new_left
                return
            else:
                return BoundaryNode.insert(root.left, left_bound, right_bound)

        if right_bound + 1 < root.value:
            new_right = BoundaryNode(root.right.value, None, None)
            new_left = BoundaryNode(root.value, root.left, new_right)
            root.value = left_bound
            root.right.value = right_bound
            root.left = new_left
            return

        if left_bound < root.value:
            root.value = left_bound

        if right_bound > root.right.value:
            root.right.value = right_bound
            while root.left is not None and root.right.value + 1 >= root.left.value:
                root.right.value = max(root.right.value, root.left.right.value)
                root.left = root.left.left

    @staticmethod
    def get_total_bounds_sum(root: "BoundaryNode", accu):
        accu += root.right.value - root.value + 1
        if root.left is None:
            return accu
        return BoundaryNode.get_total_bounds_sum(root.left, accu)
```
