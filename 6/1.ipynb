```python
INPUT="""123 328  51 64 
 45 64  387 23 
  6 98  215 314
*   +   *   +  """
```

```python
import numpy as np
```

```python
def sum_all_elements(row):
    sum = 0
    for num in row:
        sum += int(num)
    return sum

def multiply_all_elements(row):
    product = int(row[0])
    for num in row[1:]:
        product *= int(num)
    return product
```

```python
lines = INPUT.splitlines()
matrix = []

for line in lines[:-1]:
    row = []
    symbols = line.split(" ")
    for symbol in symbols:
        if symbol == "":
            continue
        row.append(symbol)
    matrix.append(row)

last_row = []
for symbol in lines[-1].split(" "):
    if symbol == "":
        continue
    last_row.append(symbol)
matrix.insert(0, last_row)

print(matrix)

np_matrix = np.matrix(matrix)

print(np_matrix.transpose())


sum = 0
for row in np_matrix.transpose():
    row = np.asarray(row).flatten()
    if row[0] == "+":
        sum += sum_all_elements(row[1:])
    elif row[0] == "*":
        sum += multiply_all_elements(row[1:])
print(sum)
```
